Data preprocessing

```python
# data preprocessing

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
import re
# Load dataset
df = pd.read_csv('edos_labelled_data.csv')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def advanced_clean_text(text):
    text = text.lower()
    text = re.sub(r"\[user\]|\[url\]", "", text)
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"#\w+", "", text)
    
    # Keep some punctuation that might be meaningful for sexism detection
    text = re.sub(r"[^a-z\s!?.]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    
    # Optional: lemmatization (can help or hurt, test it)
    # lemmatizer = WordNetLemmatizer()
    # text = " ".join([lemmatizer.lemmatize(word) for word in text.split()])
    
    return text
df["clean_text"] = df["text"].apply(advanced_clean_text)
train_df = df[df["split"] == "train"]
test_df = df[df["split"] == "test"]

y_train = train_df["label"]
y_test = test_df["label"]
```

Feature Engineering

```python
# Going to use TF-IDF vectorization for text features
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
word_vectorizer = TfidfVectorizer(
    analyzer='word',
    ngram_range=(1,3),
    max_features=20000,
    min_df=2,
    max_df=0.85,
    sublinear_tf=True
)
# Method 2: Character-level TF-IDF (CRUCIAL - this is your second method!)
char_vectorizer = TfidfVectorizer(
    analyzer='char',
    ngram_range=(3,5),
    max_features=15000,
    min_df=3,
    max_df=0.90,
    sublinear_tf=True
)

# Fit both
X_train_word = word_vectorizer.fit_transform(train_df["clean_text"])
X_test_word = word_vectorizer.transform(test_df["clean_text"])

X_train_char = char_vectorizer.fit_transform(train_df["clean_text"])
X_test_char = char_vectorizer.transform(test_df["clean_text"])

# Combine them (this will be your best feature set)
X_train_combined = hstack([X_train_word, X_train_char])
X_test_combined = hstack([X_test_word, X_test_char])

print(f"Word features: {X_train_word.shape[1]}")
print(f"Char features: {X_train_char.shape[1]}")
print(f"Combined features: {X_train_combined.shape[1]}")
```

```python

# we used AI for this part btw so will have to cite that but what it does is:
# for each minority sample, it finds k nearest neighbors and generates synthetic samples
# it will interpolate new points between knn and then it operates in the numeric feature space
# so it assumes interpolation yields meaningful examples
# and all this reduces bias towards majority class and helps improve recall and F1 for minority class
# using this helped me get to 0.81 weighted F1 score

from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42, k_neighbors=5)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_combined, y_train)

print(f"\nOriginal class distribution:")
print(y_train.value_counts())
print(f"\nBalanced class distribution:")
print(pd.Series(y_train_balanced).value_counts())
```

Model use

```python
 
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, f1_score
import xgboost as xgb
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# Encode labels for XGBoost
le = LabelEncoder()
y_train_balanced_encoded = le.fit_transform(y_train_balanced)
y_test_encoded = le.transform(y_test)

# Model 1: Logistic Regression
print("\n" + "="*50)
print("LOGISTIC REGRESSION - Combined Features + SMOTE")
print("="*50)
lr_optimized = LogisticRegression(
    max_iter=5000,
    C=3.0,  # Less regularization
    solver='saga',
    class_weight='balanced',
    random_state=42
)

lr_optimized.fit(X_train_balanced, y_train_balanced)
y_pred_opt = lr_optimized.predict(X_test_combined)

from sklearn.metrics import classification_report, f1_score
f1_opt = f1_score(y_test, y_pred_opt, average='weighted')
print(classification_report(y_test, y_pred_opt))
print(f"üéØ Optimized LR F1: {f1_opt:.4f}")


# Model 2: SVM
print("\n" + "="*50)
print("SVM - Combined Features + SMOTE")
print("="*50)
svm = LinearSVC(C=0.5, class_weight='balanced', max_iter=2000, random_state=42)
svm.fit(X_train_balanced, y_train_balanced)
y_pred_svm = svm.predict(X_test_combined)
print(classification_report(y_test, y_pred_svm))
f1_svm = f1_score(y_test, y_pred_svm, average='weighted')
print(f"üéØ Weighted F1: {f1_svm:.4f}")

# Model 3: XGBoost (usually the winner!)
print("\n" + "="*50)
print("XGBOOST - Combined Features + SMOTE")
print("="*50)
xgb_model = xgb.XGBClassifier(
    n_estimators=300,
    max_depth=7,
    learning_rate=0.1,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    eval_metric='logloss'
)
xgb_model.fit(X_train_balanced, y_train_balanced_encoded)
y_pred_xgb_encoded = xgb_model.predict(X_test_combined)
y_pred_xgb = le.inverse_transform(y_pred_xgb_encoded)
print(classification_report(y_test, y_pred_xgb))
f1_xgb = f1_score(y_test, y_pred_xgb, average='weighted')
print(f"üéØ Weighted F1: {f1_xgb:.4f}")

# ============================================
# STEP 4: Create results table for ALL combinations
# ============================================
results_list = []

# Test all feature combinations with all models
feature_sets = [
    ("Word TF-IDF", X_train_word, X_test_word, y_train, y_test),
    ("Char TF-IDF", X_train_char, X_test_char, y_train, y_test),
    ("Combined (Word+Char)", X_train_combined, X_test_combined, y_train, y_test),
    ("Combined + SMOTE", X_train_balanced, X_test_combined, y_train_balanced, y_test)
]

models = [
    ("Logistic Regression", LogisticRegression(max_iter=1000, C=1.0, random_state=42)),
    ("SVM", LinearSVC(C=0.5, max_iter=2000, random_state=42)),
]

print("\n" + "="*70)
print("GENERATING COMPLETE RESULTS TABLE")
print("="*70)

for feat_name, X_tr, X_te, y_tr, y_te in feature_sets:
    for model_name, model in models:
        try:
            model.fit(X_tr, y_tr)
            y_pred = model.predict(X_te)
            report = classification_report(y_test, y_pred, output_dict=True)
            
            results_list.append({
                "Feature + Model": f"{feat_name} + {model_name}",
                "Sexist (P)": f"{report['sexist']['precision']:.4f}",
                "Sexist (R)": f"{report['sexist']['recall']:.4f}",
                "Sexist (F1)": f"{report['sexist']['f1-score']:.4f}",
                "Not Sexist (P)": f"{report['not sexist']['precision']:.4f}",
                "Not Sexist (R)": f"{report['not sexist']['recall']:.4f}",
                "Not Sexist (F1)": f"{report['not sexist']['f1-score']:.4f}",
                "Weighted (P)": f"{report['weighted avg']['precision']:.4f}",
                "Weighted (R)": f"{report['weighted avg']['recall']:.4f}",
                "Weighted (F1)": f"{report['weighted avg']['f1-score']:.4f}"
            })
        except Exception as e:
            print(f"Error with {feat_name} + {model_name}: {e}")

# Add XGBoost separately (needs encoded labels)
for feat_name, X_tr, X_te, y_tr, y_te in feature_sets:
    try:
        y_tr_enc = le.fit_transform(y_tr) if feat_name != "Combined + SMOTE" else y_train_balanced_encoded
        xgb_temp = xgb.XGBClassifier(n_estimators=300, max_depth=7, learning_rate=0.1, random_state=42, eval_metric='logloss')
        xgb_temp.fit(X_tr, y_tr_enc)
        y_pred_enc = xgb_temp.predict(X_te)
        y_pred = le.inverse_transform(y_pred_enc)
        report = classification_report(y_test, y_pred, output_dict=True)
        
        results_list.append({
            "Feature + Model": f"{feat_name} + XGBoost",
            "Sexist (P)": f"{report['sexist']['precision']:.4f}",
            "Sexist (R)": f"{report['sexist']['recall']:.4f}",
            "Sexist (F1)": f"{report['sexist']['f1-score']:.4f}",
            "Not Sexist (P)": f"{report['not sexist']['precision']:.4f}",
            "Not Sexist (R)": f"{report['not sexist']['recall']:.4f}",
            "Not Sexist (F1)": f"{report['not sexist']['f1-score']:.4f}",
            "Weighted (P)": f"{report['weighted avg']['precision']:.4f}",
            "Weighted (R)": f"{report['weighted avg']['recall']:.4f}",
            "Weighted (F1)": f"{report['weighted avg']['f1-score']:.4f}"
        })
    except Exception as e:
        print(f"Error with {feat_name} + XGBoost: {e}")

# Create final results dataframe
results_df = pd.DataFrame(results_list)
print("\n" + "="*70)
print("FINAL RESULTS TABLE")
print("="*70)
print(results_df.to_string(index=False))

# Find best result
best_idx = results_df['Weighted (F1)'].astype(float).idxmax()
print("\n" + "="*70)
print("üèÜ BEST PERFORMANCE üèÜ")
print("="*70)
print(results_df.iloc[best_idx])
```
